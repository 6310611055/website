```python
df5days = pd.read_csv("../Dataset/five_day_activity.csv")
df5days = df5days.drop(columns=['Unnamed: 0'])
df5days
```

```python
data_5day = {'5days_sleep_time_activity':[] , '5days_day_time_activity':[], 'Id': []}
dfavg = pd.DataFrame(data_5day)
dfavg
```

```python
sleep_time_activity_5 = []
day_time_activity_5 = []
id_5 = []

def sum5days(df):
    for i in range(len(df['00:00'])):
        sum_sleeptime = 0
        sum_daytime = 0
        #print((df.iloc[i].to_numpy())[:8])
        #print((df.iloc[i].to_numpy())[8:23])
        for j in range(5):
            print((df.iloc[i].to_numpy())[(24 * j):(8 + 24 * j)])
            print((df.iloc[i].to_numpy())[-1 - (24 * j)])
            print((df.iloc[i].to_numpy())[8 + (24 * j):23 + (24 * j)])
            sum_sleeptime += sum((df.iloc[i].to_numpy())[(24 * j):(8 + 24 * j)]) + (df.iloc[i].to_numpy())[-1 - (24 * j)]
            sum_daytime += sum((df.iloc[i].to_numpy())[8 + (24 * j):23 + (24 * j)])
        #sum_sleeptime = sum((df.iloc[i].to_numpy())[:8]) + (df.iloc[i].to_numpy())[-2]
        print(sum((df.iloc[i].to_numpy())[8:23]))
        #sum_daytime = sum((df.iloc[i].to_numpy())[8:23])
        sleep_time_activity_5.append(sum_sleeptime/5)
        day_time_activity_5.append(sum_daytime/5)
        if i < 23:
            id_5.append('condition')
        else:
            id_5.append('control')
            
        #id.append((df.iloc[i].to_numpy())[-1])
        #print(sum_sleeptime)
sum5days(df5days)

data_5day['5days_sleep_time_activity'] = sleep_time_activity_5
data_5day['5days_day_time_activity'] = day_time_activity_5
data_5day['Id'] = id_5
```

```python
def classifier_modeling():
    model = Sequential()
    model.add(Dense(8, input_shape=(11,), activation='sigmoid'))
    model.add(Dropout(rate=0.05))
    model.add(BatchNormalization())
    model.add(Dense(4, activation='swish'))
    model.add(Dropout(rate=0.02))
    model.add(BatchNormalization())
    model.add(Dense(2, activation='softmax'))
    model.compile(loss='binary_crossentropy',
                  optimizer='adam', metrics=['accuracy'])
    return model
```

```python
from hyperopt.pyll.base import scope
from hyperopt import STATUS_OK
#quniform returns float, some parameters require int; use this to force int
space = {'rate1': hp.uniform('rate1', 0.1, 0.6),
         'rate2': hp.uniform('rate2', 0.1, 0.6),
        #  'rate3': hp.uniform('rate3', 0.01, 0.5),
        #  'rate4': hp.uniform('rate4', 0.01, 0.5),
        #  'rate5': hp.uniform('rate5', 0.01, 0.5),
        #  'rate6': hp.uniform('rate6', 0.01, 0.5),
         'unit1': scope.int(hp.quniform('unit1', 10, 150, 2)),
         'unit2': scope.int(hp.quniform('unit2', 10, 150, 2)),
        #  'unit3': scope.int(hp.quniform('unit3', 10, 150, 5)),
        #  'unit4': scope.int(hp.quniform('unit4', 10, 150, 5)),
        #  'unit5': scope.int(hp.quniform('unit5', 10, 150, 5)),
         'epochs': scope.int(hp.quniform('epochs', 50, 130, 1)),
         'learning_rate': hp.uniform('learning_rate', 0.0001, 0.05),
         'batch_size':
         scope.int(hp.quniform('batch_size', 1, 30, 1)),
        #  'layers': scope.int(hp.quniform('layers', 2, 3, 1)),
         'activation1': hp.choice('activation1', ['relu', 'softmax', 'sigmoid', 'softsign', 'tanh', 'selu', 'elu', 'swish']),
         'activation2': hp.choice('activation2', ['relu', 'softmax', 'sigmoid', 'softsign', 'tanh', 'selu', 'elu', 'swish']),
        #  'activation3': hp.choice('activation3', ['relu', 'softmax', 'sigmoid', 'softsign', 'tanh', 'selu', 'elu','swish']),
        #  'activation4': hp.choice('activation4', ['relu', 'softmax', 'sigmoid', 'softsign', 'tanh', 'selu', 'elu', 'swish']),
         #'activation_out': hp.choice('activation_out', ['relu', 'softmax', 'sigmoid', 'softsign', 'tanh', 'selu', 'elu', 'swish']),
         'optimizer': hp.choice('optimizer', ['Adam', 'Adamax', 'Nadam', 'SGD', 'Ftrl', 'RMSprop'])

        }
```

```python
def f_nn(params):
    
    # optimizer_selected = params['optimizer']
    optimizer = keras.optimizers.Adam(lr=params['learning_rate'])

    model = classifier_modeling()

    # First layer specifies input_shape and returns sequences
    model.add(Dense(units=params['unit1'],
                    input_shape=(x_train.shape[1],), activation=params['activation1']))
    model.add(Dropout(rate=params['rate1']))
    model.add(BatchNormalization())

    # Middle layers return sequences

    model.add(
        Dense(units=params['unit2'], activation=params['activation2']))
    model.add(Dropout(rate=params['rate2']))
    model.add(BatchNormalization())

    

    model.add(Dense(2, activation='elu'))
   
    model.compile(optimizer=optimizer, loss='mean_squared_error')

    # es = EarlyStopping(monitor='val_loss', mode='min',
    #                    verbose=1, patience=15)
    result = model.fit(x_train, y_train,
                       verbose=0,
                       validation_split=0.1,
                       batch_size=params['batch_size'],
                       epochs=params['epochs'])

    # Get the lowest validation loss of the training epochs
    validation_loss = np.amin(result.history['val_loss'])
    print('Best validation loss of epoch:', validation_loss)

    return {'loss': validation_loss,
            'status': STATUS_OK,
            'model': model,
            'params': params}
```

```python
from hyperopt import Trials

trials = Trials()
best = fmin(f_nn,
            space,
            algo=tpe.suggest,
            max_evals=50,
            trials=trials)
```

```python
best_model = trials.results[np.argmin([r['loss'] for r in 
    trials.results])]['model']
best_params = trials.results[np.argmin([r['loss'] for r in 
    trials.results])]['params']
worst_model = trials.results[np.argmax([r['loss'] for r in 
    trials.results])]['model']
worst_params = trials.results[np.argmax([r['loss'] for r in 
    trials.results])]['params']

print(best_params)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

from sklearn.metrics import RocCurveDisplay
from sklearn import preprocessing
from tensorflow import keras

from hyperopt import hp
from hyperopt import fmin, tpe
```

```python
ml_df = pd.read_csv(
    "../Dataset/for training/4_feat_wo_fill.csv")
ml_df.drop(["Unnamed: 0"], axis=1, inplace=True)
ml_df['Id'] = ['condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'condition', 'control', 'control',
               'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control', 'control']
ml_df
```

```python
x = []
y = []
count = 0
count_g = 0
count_o = 0

for i in range(len(ml_df["Id"])):

    if ml_df["Id"][i] == "condition":
        y.append([1])
        x.append([ml_df["5days_sleep_time_activity"][i],
                  ml_df["5days_day_time_activity"][i], ml_df["age"][i]])

    elif ml_df["Id"][i] == "control":
        y.append([0])
        x.append([ml_df["5days_sleep_time_activity"][i],
                  ml_df["5days_day_time_activity"][i], ml_df["age"][i]])


print(f'Select: {len(y)} set')
y = np.array(y)
x = np.array(x)

```

```python
# Add noisy features
random_state = np.random.RandomState(0)


# Train Test split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=random_state)
```

```python
def classifier_modeling():
    model = Sequential()
    model.add(Dense(8, input_shape=(3,), activation='relu'))
    model.add(Dense(4, activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',
                  optimizer='adam', metrics=['accuracy'])
    return model
```

```python
classifier_modeling()
```

```python
from hyperopt.pyll.base import scope
from hyperopt import STATUS_OK
#quniform returns float, some parameters require int; use this to force int
space = {'rate': hp.uniform('rate', 0.01, 0.5),
         'unit1': scope.int(hp.quniform('unit1', 10, 200, 10)),
         'unit2': scope.int(hp.quniform('unit2', 10, 200, 10)),
         'unit3': scope.int(hp.quniform('unit3', 10, 200, 10)),
         'batch_size':
         scope.int(hp.quniform('batch_size', 100, 250, 25)),
         'layers': scope.int(hp.quniform('layers', 1, 3, 1)),
         'activation': hp.choice('activation', ['relu', 'softmax', 'sigmoid']),
        'optimizer': hp.choice('optimizer', ['adam', 'adadelta', 'sgd', 'RMSprop'])

        }
```

```python
def f_nn(params):

    model = classifier_modeling()

    if params['layers'] == 1:
        model.add(Dense(units=params['unit1'],
                        input_shape=(x_train.shape[1],), activation=params['activation']))

        model.add(Dropout(rate=params['rate']))
    else:
        # First layer specifies input_shape and returns sequences
        model.add(Dense(units=params['unit1'],
                        input_shape=(x_train.shape[1],), activation=params['activation']))
        model.add(Dropout(rate=params['rate']))

        # Middle layers return sequences
        for i in range(params['layers']-1):
            model.add(
                Dense(units=params['unit'+str(i+2)], activation=params['activation']))
        model.add(Dropout(rate=params['rate']))

    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer=params['optimizer'], loss='mean_squared_error')

    # es = EarlyStopping(monitor='val_loss', mode='min',
    #                    verbose=1, patience=15)
    result = model.fit(x_train, y_train,
                       verbose=0,
                       validation_split=0.1,
                       batch_size=params['batch_size'],
                       epochs=200)

    # Get the lowest validation loss of the training epochs
    validation_loss = np.amin(result.history['val_loss'])
    print('Best validation loss of epoch:', validation_loss)

    return {'loss': validation_loss,
            'status': STATUS_OK,
            'model': model,
            'params': params}
```

```python
from hyperopt import Trials

trials = Trials()
best = fmin(f_nn,
            space,
            algo=tpe.suggest,
            max_evals=5,
            trials=trials)
```

```python
best_model = trials.results[np.argmin([r['loss'] for r in 
    trials.results])]['model']
best_params = trials.results[np.argmin([r['loss'] for r in 
    trials.results])]['params']
worst_model = trials.results[np.argmax([r['loss'] for r in 
    trials.results])]['model']
worst_params = trials.results[np.argmax([r['loss'] for r in 
    trials.results])]['params']

print(best_params)
```
